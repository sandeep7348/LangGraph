```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
class BatsmanData(TypedDict):
    name:str
    runs:int
    balls:int
    fours:int
    sixes:int

    sr:float
    bpb:float
    boundary_percentage:float
    summary:str
graph=StateGraph(BatsmanData)

def calculate_strike_rate(state:BatsmanData):
    sr=state['runs']/state['balls']*100
    return {'sr':sr}
def calulate_bpb(state:BatsmanData):
    bpb=state['balls']/state['runs']
    return {'bpb':bpb}
def calculate_boundary_percentage(state:BatsmanData):
    boundary_percentage=(state['fours']+state['sixes'])/state['balls']*100
    return {'boundary_percentage':boundary_percentage}

def summary(state:BatsmanData):
    summary=f"{state['name']} scored {state['runs']} runs in {state['balls']} balls with a strike rate of {state['sr']:.2f} and a boundary percentage of {state['boundary_percentage']:.2f}%."
    return {'summary':summary}
graph.add_node('calculate_strike_rate',calculate_strike_rate)
graph.add_node('calculate_bpb',calulate_bpb)
graph.add_node('calculate_boundary_percentage',calculate_boundary_percentage)
graph.add_node('summary',summary)
graph.add_edge(START,'calculate_strike_rate')
graph.add_edge(START,'calculate_bpb')
graph.add_edge(START,'calculate_boundary_percentage')
graph.add_edge('calculate_strike_rate','summary')
graph.add_edge('calculate_boundary_percentage','summary')
graph.add_edge('calculate_bpb','summary')
input_data=BatsmanData(name='Vaibhav',runs=75,balls=25,fours=4,sixes=6)
workflow=graph.compile()
result=workflow.invoke(input_data)
print(result)
```
