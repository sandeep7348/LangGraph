```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict,Literal

class Roots(TypedDict):
    a: int
    b: int
    c: int
    equation: str
    discriminant: int
    result: str

graph = StateGraph(Roots)

def show_equation(state: Roots):
    return {
        **state,
        "equation": f"{state['a']}x^2+{state['b']}x+{state['c']}=0"
    }

def calculate_discriminant(state: Roots) -> Roots:
    d = state["b"]**2 - 4 * state["a"] * state["c"]
    return {"discriminant":d}
def checkCondition(state:Roots)->Literal["two_real_roots","one_real_root","no_real_root"]:
    if state["discriminant"] > 0:
        return "two_real_roots"
    elif state["discriminant"] == 0:
        return "one_real_root"
    else:
        return "no_real_root"

def two_real_roots(state: Roots) -> Roots:
    d = state["discriminant"]
    a = state["a"]
    b = state["b"]
    root1 = (-b + d**0.5) / (2*a)
    root2 = (-b - d**0.5) / (2*a)
    return {"result": f"Two real roots: {root1} and {root2}"}
def one_real_root(state: Roots) -> Roots:
    a = state["a"]
    b = state["b"]
    root = -b / (2*a)
    return {"result": f"One real root: {root}"}
def no_real_root(state: Roots) -> Roots:
    return {"result": "No real roots"}
graph.add_node("show_equation", show_equation)
graph.add_node("calculate_discriminant", calculate_discriminant)
graph.add_node("two_real_roots", two_real_roots)
graph.add_node("one_real_root", one_real_root)
graph.add_node("no_real_root", no_real_root)

graph.add_edge(START, "show_equation")
graph.add_edge("show_equation", "calculate_discriminant")
graph.add_conditional_edges("calculate_discriminant",checkCondition )
graph.add_edge("two_real_roots", END)
graph.add_edge("one_real_root", END)
graph.add_edge("no_real_root", END)



workflow = graph.compile()

initial_state = Roots(
    a=1,
    b=-3,
    c=2,
    equation="",
    discriminant=0,
    result=""
)

print(workflow.invoke(initial_state))


from IPython.display import Image

Image(workflow.get_graph().draw_mermaid_png())
```
